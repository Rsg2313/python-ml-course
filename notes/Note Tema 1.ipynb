# Cargar datos a traves de la funcion Open

```python
dataframe3 = open(r"C:\Users\dsg_s\Documents\Machine Learning\python-ml-course\datasets\customer-churn-model\customer churn model.txt","r")
```

```python
cols = dataframe3.readline().strip().split(",")
n_cols = len(cols)
```

```python
counter = 0
main_dict = {}
for col in cols:
    main_dict[col] = []
```

```python
for line in dataframe3:
    values = line.strip().split(",")
    for i in range(len(cols)):
        main_dict[cols[i]].append(values[i])
    counter += 1

print ("El dataset tiene %d filas y %d columnas" % (counter, n_cols))
```

```python
df3 = pd.DataFrame(main_dict)
df3.head()
```

# Lectura y escritura de ficheros

```python
infile = r"C:\Users\dsg_s\Documents\Machine Learning\python-ml-course\datasets\customer-churn-model\customer churn model.txt"
outfile = r"C:\Users\dsg_s\Documents\Machine Learning\python-ml-course\datasets\customer-churn-model\SPACE customer churn model.txt"
```

```python
with open(infile,"r") as  infile1:
    with open(outfile, "w") as outfile1:
        for line in infile1:
            fields = line.strip().split(",")
            outfile1.write("\t".join(fields))
            outfile1.write("\n")
```
